# **Analyze and Share**


The sleepDay column was separated into the Date and Time columns, with the Time column being later removed for covenance purposes.

```r
sleepDay <- sleepDay %>% separate(SleepDay, c("Date", "Time"), sep=" ")
sleepDay$Time <- NULL
glimpse(sleepDay)
```

A new column was created for the Total Hours Asleep

```r
sleepDay <- mutate(sleepDay, TotalHoursAsleep = TotalMinutesAsleep/60)
glimpse(sleepDay)
```

Three different groups were created base on the amount of hours a person sleeps per night.
Each entry was grouped into either not enough sleep, enough sleep, or too much sleep based if they were below, inbetween, or above the recommended 7-9 hours of sleep.

```r
sleepDay <- sleepDay %>%
  mutate(sleepStatus = case_when(
    TotalHoursAsleep < 7 ~ 'Not_enough_sleep',
    TotalHoursAsleep >= 7 & TotalHoursAsleep <= 9 ~ 'Enough_sleep',
    TotalHoursAsleep > 9 ~ 'Too_much_sleep'))
```

A new table was created that contains the percentage of each group from sleepDay.

```r
Sleep_percentage <- sleepDay %>%
  group_by(sleepStatus) %>%
  summarise(number_of_users=n()) %>%
  mutate(sleepStatus_percentage = percent(c(number_of_users/sum(number_of_users))))
view(Sleep_percentage)
```

**Percentage of Sleep**

A graph showing the proportions of the three groups was created.

```r
ggplot(Sleep_percentage, aes(x="", y=sleepStatus_percentage, fill=sleepStatus))+
  geom_bar(width=1, stat="identity", color="black")+
  coord_polar("y", start = 0)+
  geom_label(aes(label = sleepStatus_percentage),
             position = position_stack(vjust = 0.5),
             show.legend = FALSE) +
  labs(title = "Percentage of Sleep") +
  guides(fill = guide_legend(title=NULL)) +
  theme_void() +
  scale_fill_discrete(labels = c("Enough Sleep", "Not Enough Sleep", "Too Much Sleep")) +
  theme(legend.position = "bottom")
```

# Insight: 
Only 46.34% of the participates over the month that they were monitored were getting enought sleep.  More than half of the time, the participates either overslept or underslept.



New columns were created for days of the week.

```r
sleepDay$NewDate <- strptime(as.character(sleepDay$Date), "%m/%d/%Y")
sleepDay$BetterDate <- format(sleepDay$NewDate, "%Y-%m-%d")
sleepDay$weekday <- wday(sleepDay$BetterDate, label=TRUE, abbr=FALSE)
```

A new dataset was created for the average hours asleep per day of the week.

```r
Average_Sleep_per_Weekday <- sleepDay %>%
  group_by(weekday) %>%
  summarise(Average_Sleep_By_Day = mean(TotalHoursAsleep))  
View(Average_Sleep_per_Weekday)
```

**Hours Asleep per day of the week**

A boxplot graph was created to show the ranges of the amount of sleep a person get per day of the week.

```r
ggplot(data = sleepDay, aes(x=weekday, y=TotalHoursAsleep, fill = weekday)) + 
  geom_boxplot() +
  labs(title = "Range of Sleep in Hours based on Day of the Week") +
  xlab(label = "Day of the Week") +
  ylab(label = "Total Hours Asleep (Hrs)") +
  scale_fill_manual(values = c("Sunday" = "red",
                               "Monday" = "orange",
                               "Tuesday" = "#1b98e0",
                               "Wednesday" = "grey",
                               "Thursday" = "purple",
                               "Friday" = "pink",
                               "Saturday" = "green"))
```

# Insight:

The graph shows that Saturdays and Sundays have the biggest fluxuations and that Wednesdays and Thursdays have the least fluxuations for the amount of sleep a person gets.

**Average Sleep per day of the week**

A bar graph was created to show the average amount of sleep a person gets per day of the week.

```r
ggplot(data = Average_Sleep_per_Weekday, aes(x=weekday, y=Average_Sleep_By_Day, fill = weekday)) + 
  geom_bar(stat = "identity") +
  labs(title = "Average Sleep in Hours based on Day of the Week") +
  xlab(label = "Day of the Week") +
  ylab(label = "Average Time Asleep (Hrs)") +
  scale_fill_manual(values = c("Sunday" = "red",
                               "Monday" = "orange",
                               "Tuesday" = "#1b98e0",
                               "Wednesday" = "grey",
                               "Thursday" = "purple",
                               "Friday" = "pink",
                               "Saturday" = "green")) +
  geom_label(aes(label = sprintf("%0.2f", round(Average_Sleep_By_Day, digits = 2))),
             position = position_stack(vjust = 0.9),
             show.legend = FALSE)
```

The graph shows that the on average, people on Sundays and Wednesday get enough of the recommended 7 to 9 hours of sleep, with the rest of the days falling below the 7 hour mark.
